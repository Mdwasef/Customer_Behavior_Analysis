```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':"purchase_amount"})
df.columns
```

```python
## create a new column age_group
# here we grouped peoples based on age like [young adult, adult, middle aged, senior]
```

```python
df['age']
```

```python
labels=['Young Adult','Adult','Middle-aged','Senior']
df['age_group']=pd.cut(df['age'],bins=[18,25,40,55,71],labels=labels)
```

```python
df.head()
```

```python
df[['age','age_group']].head()
```

```python
### create column purchase frequency days
# 'frequency_of_purchases' is in text form . so convert it into numerical values like weekly means 7, montly-> 30

df['frequency_of_purchases'].unique()

frequency_mapping={
'Fortnightly':14,
'Weekly':7,
'Annually':365,
'Quarterly':90,
'Bi-Weekly':14,
'Monthly':30,
'Every 3 Months':30
}


df['frequency_mapping']=df['frequency_of_purchases'].map(frequency_mapping)
```
